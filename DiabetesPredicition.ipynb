#Lets work on our model now. We have an idea of what is the best model to use.


```python
df = df.drop_duplicates()
```

 Train Test Split

```python
X = df.drop(columns=['Outcome'], axis = 1)
y = df['Outcome']
```

```python
# split X and y into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=16)
```

Model Training Logisitc Regression

```python
# import the class
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

classifier = svm.SVC(kernel='linear')


# fit the model with data
classifier.fit(X_train, y_train)


SVC(kernel='linear')
```

ModelEvaluation

```python
# import the necessary library
from sklearn.metrics import accuracy_score

# accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)


# accuracy score on the test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)

print('Accuracy on test data : ', test_data_accuracy)

print('Accuracy on training data : ', training_data_accuracy)

```

Making a predicitive system

```python

import warnings
from sklearn.exceptions import DataConversionWarning
warnings.filterwarnings(action='ignore', category=DataConversionWarning)

input_data =(11,143,94,33,146,36.6,0.254,51)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = classifier.predict(input_reshaped)
print(prediction)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')

```
